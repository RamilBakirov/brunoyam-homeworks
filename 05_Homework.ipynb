# Easy


```python
# Вычислил вероятность выживания, если пассажиру до 25, от 25 до 50, от 50 лет.
youth_surv = len(df.loc[((df['Age'] < 25) & (df['Survived'] == 1))])
adult_surv = len(df.loc[((df['Age'] >= 25) & (df['Survived'] == 1))]) - len(df.loc[((df['Age'] >= 50) & (df['Survived'] == 1))])
old_surv = len(df.loc[((df['Age'] >= 50) & (df['Survived'] == 1))])
total_youth = len(df.loc[(df['Age'] < 25)])
total_adult = len(df.loc[(df['Age'] >= 25)]) - len(df.loc[(df['Age'] > 50)])
total_old = len(df.loc[(df['Age'] >= 50)])
prob_youth = youth_surv / total_youth # вероятность исхода
prob_adult = adult_surv / total_adult # вероятность исхода
prob_old = old_surv / total_old # вероятность исхода
print("Вероятность выживания если тебе от 0 до 25 лет:", round(prob_youth*100, 2),'%')
print("Вероятность выживания если тебе от 25 до 50 лет:", round(prob_adult*100, 2),'%')
print("Вероятность выживания если ты старше 50 лет:", round(prob_old*100, 2),'%')
```

# Normal

```python
n = len(df.loc[(df['Sex'] == 'female')]) | len(df.loc[(df['Sex'] == 'male')]) # количество пассажиров (все мужчины и женщины из таблицы данных).
n_surv = len(df.loc[(df['Survived'] == 1)]) # количество пассажиров, тех кто выжил.
females = len(df.loc[(df['Sex'] == 'female')]) # общее количество женщин
females_surv = len(df.loc[((df['Sex'] == 'female') & (df['Survived'] == 1))]) # количетсво выживших женщин
p = ((females_surv / females) * (females / n)) / (n_surv / n)
print("Вероятность выжить женщине по теореме Байеса равна:", round(p*100, 2), '%')
```
